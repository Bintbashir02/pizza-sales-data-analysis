# Clean the data if necessary (handle missing values , incorrect data, etc)

```python
df.isnull().sum()
```

# Convert date and time to appropriate datetime format

```python
df['date'] = pd.to_datetime(df['date'])
df['time'] = pd.to_datetime(df['time'])
```

```python
#Extract day of week, hour from datetime for analysis
df['Day_of_week'] = df['date'].dt.day_name()
df['Hour'] = df['time'].dt.hour
```

```python
df
```

```python
# Total sales per order
```

```python
df['Total_sales'] = df['quantity'] * df['price']
```

```python
df
```

# Key Insights from Analysis

```python
# How many customers do we have each day?
```

```python
customer = df.groupby(df['Day_of_week'])['order_details_id'].count()
```

```python
customer
```

```python
#plot sale trends by day of the week
plt.figure(figsize=(10,6))
sns.barplot(x='Day_of_week', y='Total_sales', data=df, estimator=sum, errorbar=None)
plt.title('Total sales by day of the week')
plt.show()
```

```python
#plot sale trends by hour
plt.figure(figsize=(10,6))
sns.barplot(x='Hour', y='Total_sales', data=df, estimator=sum)
plt.title('Total sales by Hour of the day')
plt.show()
```

```python
# identify best and worst selling pizzas
```

```python
best_selling = df.groupby('category')['Total_sales'].sum().sort_values(ascending=False).head(5)
worst_selling = df.groupby('category')['Total_sales'].sum().sort_values().head(5)
```

```python
print("Best Selling Pizzas:")
print(best_selling)
print("Worst Selling Pizzas:")
print(worst_selling)
```

```python
plt.figure(figsize=(8,8))
best_selling.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Best- Selling Pizzas (by Quantity')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
